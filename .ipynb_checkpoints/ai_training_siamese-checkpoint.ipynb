```python
import chess

import chess.svg
from IPython.display import SVG, display

def visualize(fen_str):
    board = chess.Board(fen_str)
    display(chess.svg.board(board=board,size=200))

pieces = {
	'p': 1,
	'P': -1,
	'n': 2,
	'N': -2,
	'b': 3,
	'B': -3,
	'r': 4,
	'R': -4,
	'q': 5,
	'Q': -5,
	'k': 6,
	'K': -6
		}

reverse_table = dict(zip(pieces.values(),pieces.keys()))
print(reverse_table)

# print(chess.Piece.from_symbol('p'))

board = chess.Board()
board.clear_board()
board.set_piece_at(0,chess.Piece.from_symbol('p')) # indices start at a1 (0), so we need to flip coordinates
board.set_piece_at(1,chess.Piece.from_symbol('P'))

print(board)

# converts prettified board to a python-chess board
def reverse_beautify(arr):
    board = chess.Board()
    board.clear_board()
    for i,piece in enumerate(arr):
        if i<64 and piece != 0:
            board.set_piece_at(63-i,chess.Piece.from_symbol(reverse_table[piece]))
    if arr[64] == -7:
        board.turn = chess.WHITE
    else:
        board.turn = chess.BLACK
    return board
        
for i in range(5):
#     print(X1_fen[i])
    board = reverse_beautify(X1_fen[i])
    display(chess.svg.board(board=board,size=200))

# list(map(bitifyFEN,X1_fen))

# f = np.vectorize(bitifyFEN)
# print(f(X1_fen)).shape
```

```python
# training procedure

lr_schedule = tf.keras.optimizers.schedules.InverseTimeDecay(
  0.001,
  decay_steps=STEPS_PER_EPOCH*1000,
  decay_rate=1,
  staircase=False)

def get_optimizer():
  return tf.keras.optimizers.Adam(lr_schedule)

# callbacks during training: performance, early stopping, tensorboard, model checkpoints
def get_callbacks(name):
    
  return [
    tfdocs.modeling.EpochDots(),
    tf.keras.callbacks.EarlyStopping(monitor='val_binary_crossentropy', patience=2),
    tf.keras.callbacks.TensorBoard(log_dir=logdir/name),
    tf.keras.callbacks.ModelCheckpoint(filepath=checkpoint_path,
                                                 save_weights_only=True,
                                                 verbose=1)
  ]

def compile_and_fit(model, name, optimizer=None, max_epochs=100):
  if optimizer is None:
    optimizer = get_optimizer()
  model.compile(optimizer=optimizer,
                loss='binary_crossentropy',
                metrics=['accuracy', 'binary_crossentropy'])

  model.summary()

  history = model.fit(
    x=X_train,
    y=y_train,
    steps_per_epoch = STEPS_PER_EPOCH,
    epochs=max_epochs,
    validation_data=(X_test,y_test),
    callbacks=get_callbacks(name),
    verbose=2)
  return history
```

```python
from tensorflow.keras import Input

# model

# single-board input model
# model = tf.keras.models.Sequential([
#     tf.keras.layers.Dense(100, activation='relu',input_shape=(N_INPUT,)),
#     tf.keras.layers.Dense(100, activation='relu'),
#     tf.keras.layers.Dense(100, activation='relu'),
#     tf.keras.layers.Dense(2, activation='softmax')
# ])

def create_siamese_model(input_shape=N_INPUT):
    
    # Define the tensors for the two input boards
    left_input = Input(input_shape,name='board_1')
    right_input = Input(input_shape,name='board_2')

    model = tf.keras.models.Sequential([
        tf.keras.layers.Dense(100, activation='relu'),
#         tf.keras.layers.Dense(100, activation='relu'),
#         tf.keras.layers.Dense(100, activation='relu'),
    ])

    # Generate the encodings (feature vectors) for the two inputs
    encoded_l = model(left_input)
    encoded_r = model(right_input)
    
    concat = tf.keras.layers.concatenate([encoded_l,encoded_r])
    x = tf.keras.layers.Dense(100, activation='relu')(concat)
    out = tf.keras.layers.Dense(2, activation='softmax')(x)
    
    # Connect the inputs with the outputs
    net = tf.keras.Model(inputs=[left_input,right_input],outputs=out)
    
    # return the model
    return net

model = create_siamese_model()
model.summary()
# tf.keras.utils.plot_model(model, 'siamese.png')
```
